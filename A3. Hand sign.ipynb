### Imports

```python
import cv2 
import os, glob

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

image_dir = "images/hand_gestures_digits_0_5"
```

### Plot all the base images in a subplot of 2 rows and 3 columns

```python
## Code to plot the base images.
images = []
for image in glob.glob(os.path.join(image_dir, "*")):
    im = cv2.imread(image)
    im = cv2.resize(im, (375, 550))
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    images.append(im)

f, ((zero, one, two), (three, four, five)) = plt.subplots(2,3, figsize=(20, 10))
zero.imshow(images[0])
zero.set_title(f"{str(0)}")
one.imshow(images[1])
one.set_title(f"{str(1)}")
two.imshow(images[2])
two.set_title(f"{str(2)}")
three.imshow(images[3])
three.set_title(f"{str(3)}")
four.imshow(images[4])
four.set_title(f"{str(4)}")
five.imshow(images[5])
five.set_title(f"{str(5)}")
```

### Process all the images. Add all the code cells you may require.

```python
## Image processing.
dilation_kernel = np.ones((2,2), np.uint8)

grey_images = []
for image in images:
    im = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    im = cv2.GaussianBlur(im, (9,9), 0)
    im = cv2.Canny(im, 0, 52)
    im = cv2.dilate(im, dilation_kernel, iterations = 1)
    grey_images.append(im)
```

### Plot all the processed images in a subplot of 2 rows and 3 columns.

```python
f, ((zero, one, two), (three, four, five)) = plt.subplots(2,3, figsize=(20, 10))
zero.imshow(grey_images[0],cmap='gray')
zero.set_title(f"{str(0)}")
one.imshow(grey_images[1],cmap='gray')
one.set_title(f"{str(1)}")
two.imshow(grey_images[2],cmap='gray')
two.set_title(f"{str(2)}")
three.imshow(grey_images[3],cmap='gray')
three.set_title(f"{str(3)}")
four.imshow(grey_images[4],cmap='gray')
four.set_title(f"{str(4)}")
five.imshow(grey_images[5],cmap='gray')
five.set_title(f"{str(5)}")
```
